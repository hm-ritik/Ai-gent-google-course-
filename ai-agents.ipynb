```python
from google.adk.agents import Agent
from google.adk.runners import InMemoryRunner
from google.adk import tools   # import the tools module itself

# Create your agent (note: tool added by string name)
class SearchAgent(Agent):
    def __init__(self):
        super().__init__(
            name="SearchAgent",
            model="gemini-1.5-flash",
            tools=[tools.google_search],   # ‚úÖ use module reference, not class
        )

# Create the runner
agenti = SearchAgent()
runner = InMemoryRunner(app_name="MyFirstAgentApp", agent=agenti)

print("‚úÖ Agent created successfully!")


```

```python
# ===== GET AI RESPONSE =====
# Simple, direct approach using genai

prompt = input("\nü§ñ Enter your Query: ")

try:
    # Use gemini-2.5-flash (fast and reliable)
    model = genai.GenerativeModel('gemini-2.5-flash')
    response = model.generate_content(prompt)
    
    print("\n" + "="*70)
    print("‚úÖ AI Agent Response:")
    print("="*70)
    print(response.text)
    print("="*70)
    
except Exception as e:
    print(f"\n‚ùå Error: {e}")
    print(f"Trying alternative model...\n")
    
    try:
        # Fallback to gemini-2.0-flash
        model = genai.GenerativeModel('gemini-2.0-flash')
        response = model.generate_content(prompt)
        
        print("\n" + "="*70)
        print("‚úÖ AI Agent Response (using gemini-2.0-flash):")
        print("="*70)
        print(response.text)
        print("="*70)
        
    except Exception as e2:
        print(f"\n‚ùå Failed: {e2}")
```
